```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
def extract_amplitude_from_wav(file_path):
    # WAV 파일 읽기
    y, sr = librosa.load(file_path)

    # amplitude 추출
    amplitude = librosa.feature.rms(y=y, frame_length=1024, hop_length=256)

    return amplitude, sr, len(y)
```

```python
# WAV 파일 경로
file_path = "/content/drive/MyDrive/전체 데이터셋.wav"
```

```python
# amplitude 추출
amplitude, sr, num_samples = extract_amplitude_from_wav(file_path)
```

```python
# 결과 출력
print("Amplitude:", amplitude)
```

```python
# 시간 벡터 생성
time = np.linspace(0, num_samples / sr, num=amplitude.shape[1])
```

```python
# 시각화
plt.figure(figsize=(10, 4))
plt.plot(time, amplitude[0], color='b')
plt.title('Amplitude over Time')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```
